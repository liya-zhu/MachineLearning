```python
#Using the sequential function np.max and broadcasting, n
#ormalize images such that the largest value within each color-channel of each image is 1.

images = np.random.rand(500, 48, 48, 3)
```

```python
norm=images/np.max(images,axis=(1,2)).reshape(500,1,1,3)
```

```python
np.all(norm.max(axis=(1,2))==1)
```
